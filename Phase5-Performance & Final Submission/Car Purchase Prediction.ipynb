```python
data_car = data.drop(columns=['User ID'], axis=0, inplace=False)
```

```python
data_car.head()
```

# Encoding Categorical column

```python
#Performing Train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_car.iloc[:, 0:3], data_car.iloc[:, -1], test_size=0.2, random_state=42)
```

```python
X_train.head(), X_test.head(), y_train.head(), y_test.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder( drop='first', sparse=False, dtype=np.int32)
```

```python
X_train_new = ohe.fit_transform(X_train[['Gender']])

X_test_new = ohe.transform(X_test[['Gender']])
```

```python
X_train = pd.DataFrame(np.hstack((X_train[['Age','AnnualSalary']].values, X_train_new)), columns=['Age', 'AnnualSalary', 'Gender'])
X_test= pd.DataFrame(np.hstack((X_test[['Age','AnnualSalary']].values, X_test_new)),columns=['Age', 'AnnualSalary', 'Gender'])
```

```python
X_train.head()
# X_test.head()
```

# Scaling of Data columns

**We observe that since, Age and AnnualSalary columns have a very different scale, while building the model this can be a problem. Hence we will scale the data in these columns using standard scaler.**

```python
# Importing classes

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# Scaling Age and AnnualSalary

X_train_Scaled = scaler.fit_transform(X_train[['Age', 'AnnualSalary']])
X_test_Scaled = scaler.transform(X_test[['Age', 'AnnualSalary']])
```

```python
X_train = pd.DataFrame(np.hstack(( X_train_Scaled, X_train[['Gender']].values)), columns=['Age', 'AnnualSalary', 'Gender'])
X_test= pd.DataFrame(np.hstack(( X_test_Scaled, X_test[['Gender']].values)),columns=['Age', 'AnnualSalary', 'Gender'])
```

```python
X_train['Gender'] = X_train['Gender'].astype(int)
X_test['Gender'] = X_test['Gender'].astype(int)
```

```python
X_train.head()
```

# Model Building

**We clearly see that the problem we have is a classification problem. Hence, we will try to apply all the classification based algorithm to check which algorithm gives best accuracy.**

# Logistic Regression

```python
sns.scatterplot(data['Age'], data['AnnualSalary'], hue=data['Purchased'], style=data['Gender'])
```

**We observe that the data here is not linearly seperable. Hence, we see that logistic regression may not be the best technique in this case. Still we will try to apply logistic regression and get the best result.**

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1,10):
    km= KMeans(n_clusters=i)
    km.fit_predict(X_train)
    wcss.append(km.inertia_)

wcss
```

```python
plt.plot(range(1,10),wcss,color='r')
plt.title('Elbow Method Graph')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier 
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# Make predictions on the test set
y_pred = logreg.predict(X_test)

# Calculate accuracy on the test set
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
naive_bayes = GaussianNB()
naive_bayes.fit(X_train, y_train)

y_pred = naive_bayes.predict(X_test)
# Calculate accuracy on the test set
accuracy1 = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy1)
```

```python
svm_classifier = SVC(kernel='linear') 
svm_classifier.fit(X_train, y_train)

y_pred = svm_classifier.predict(X_test)

# Calculate accuracy on the test set
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Decision Tree



classifier= DecisionTreeClassifier(criterion='entropy', random_state=0)  
classifier.fit(X_train, y_train)

y_pred= classifier.predict(X_test)  

from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test, y_pred) 
cm

from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
from sklearn.neighbors import KNeighborsClassifier 

k_classifier= KNeighborsClassifier(n_neighbors=3, metric='minkowski', p=2 )  
k_classifier.fit(X_train, y_train)  

#Predicting the test set result  
y_pred= k_classifier.predict(X_test) 
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Define the hyperparameters and their possible values for tuning
param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

# Create the KNeighborsClassifier model
k_classifier = KNeighborsClassifier()

# Perform Grid Search with cross-validation
grid_search = GridSearchCV(estimator=k_classifier, param_grid=param_grid, scoring='accuracy', cv=5)
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


best_k_classifier = KNeighborsClassifier(n_neighbors=best_params['n_neighbors'], metric=best_params['metric'])
best_k_classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = best_k_classifier.predict(X_test)

# Calculate accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy)
```
